```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x =np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Functional Method
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented API
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title("Reverse")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
axes[0].plot(x,y)
axes[0].set_title("1st plot")
axes[1].plot(y,x)
axes[1].set_title("2nd plot")
```

```python
fig=plt.figure(figsize=(12,6),dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(figsize=(6,6),dpi=100, nrows=2, ncols=1)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my plot.png')
```

```python
fig=plt.figure(figsize=(6,6),dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X squared')
ax.plot(x,x**3, label = 'X cubic')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#38CABE', lw=3,ls='--', alpha=0.6, marker='o', ms=10,
       mfc='yellow',mew=3,mec='blue')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
