```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill',data=tips,estimator=np.std, hue='day')
plt.legend(loc='upper right')
```

```python
sns.countplot(x='sex', data=tips, hue='day')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split= True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex',dodge= True)
plt.legend(loc='upper left')
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, dodge=True, color='black')
sns.violinplot(x='day', y='total_bill', data=tips, split= True)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, hue='sex', kind='swarm')
```
