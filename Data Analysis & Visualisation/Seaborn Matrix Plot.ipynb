```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
tc=tips.corr()
```

```python
sns.heatmap(tc, annot=True,cmap='coolwarm')
```

```python
fp=flights.pivot_table(index='month',values='passengers',columns='year')
```

```python
sns.heatmap(fp,linecolor='blue', linewidth=1)
```

```python
sns.clustermap(fp,linecolor='white', linewidth=1)
```

```python
sns.clustermap(fp,linecolor='white', linewidth=1,standard_scale=1)
```
